```python
# Defined for easy access
colorPalette = sns.color_palette()
snsBlue = colorPalette[0]
snsGreen = colorPalette[1]
snsRed = colorPalette[2]
snsPurple = colorPalette[3]
snsYellow = colorPalette[4]
snsAzure = colorPalette[5]
```

```python
labels = [183, 1467, 12029, 32408, 66602]
label_names = ['1 review', '10 reviews', '100 reviews', '250 reviews', '500 reviews']
```

```python
words_per_sec_serial = [43.36, 39.76, 37.16, 41.20, 40.11]
```

```python
words_per_sec_v1 = [3.95, 34.28, 65.75, 66.39, 69.22]
```

```python
plt.plot(labels, words_per_sec_serial, color=snsBlue, label='Serial')
plt.plot(labels, words_per_sec_v1, color=snsRed, label='SPARK')
plt.title('Context-level spell-checking')
plt.ylabel('Words per second')
plt.xticks(labels, label_names, rotation='vertical')
plt.xlabel('Number of words')
plt.xlim(xmin=min(labels), xmax=max(labels))
plt.show()
```
